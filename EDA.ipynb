Can you create a plot showing the growth rate of new customers over the months?

```python
group6=df.groupby('Customer_ID')
first_order=group6['Order_Date'].agg(np.min)
```

```python
first_order=pd.DataFrame(first_order)
```

```python
first_order['Customer_ID']=first_order.index
```

```python
first_order['Order_Date']= pd.to_datetime(first_order['Order_Date'],format='%Y-%m-%d')
```

```python
first_order['month']=first_order['Order_Date'].dt.month
first_order['year']=first_order['Order_Date'].dt.year
```

```python
group7=first_order.groupby(['month','year'])
re=group7['Customer_ID'].agg('count')
```

```python
res=pd.DataFrame(re)
res['month']=res.index.get_level_values(0)
res['year']=res.index.get_level_values(1)
```

```python
res_ordered= res.reset_index(drop=True)
```

```python
res_ordered= res_ordered.sort_values(['year','month'])
```

```python
res_ordered['year']= res_ordered['year'].astype(str)
res_ordered['month']= res_ordered['month'].astype(str)
```

```python
import datetime
res_ordered['Date']=pd.to_datetime(res_ordered['year']+ '-'+res_ordered['month'])
```

```python
res_ordered["GrowthRate"] = res_ordered["Customer_ID"].pct_change().round(2)
```

```python
fig=plot.figure(figsize=(12,6))
sn.lineplot(res_ordered['Date'],res_ordered['GrowthRate'])
```
