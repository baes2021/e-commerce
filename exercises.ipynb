```python
import sqlite3
import pandas as pd
db = sqlite3.connect('Samp.db')
```

```python
#--1 What is the category generating the maximum sales revenue?

new_cur = db.cursor()
res = new_cur.execute("Select *, round(sum(sales),2) as sales_revenue \
                      from Product \
                      join Sale \
                      on sale.Product_ID=Product.Product_ID \
                      group by Category \
                      order by sales_revenue DESC")
ans = res.fetchall()
df=[]
for i in ans:
    df.append(i)
dff1=pd.DataFrame(df,columns=['Product_ID','Product_Name','Category','Sub_Category', 'Product-ID','Quantity','Discount','Profit','Sales','Customer_ID','Sales_revenue'])
dff1
```

```python
#-- 3 Are they making a loss in any categroies?

new_cur = db.cursor()
res = new_cur.execute("select *, round(sum(Profit),2) as sales_revenue \
                      from Product \
                      join Sale \
                      on sale.Product_ID=Product.Product_ID \
                      group by Category \
                      order by sales_revenue DESC")
ans = res.fetchall()
df=[]
for i in ans:
    df.append(i)
dff2=pd.DataFrame(df,columns=['Product_ID','Product_Name','Category','Sub_Category', 'Product-ID','Quantity','Discount','Profit','Sales','Customer_ID','Sales_revenue'])
dff2
```

```python
# 4 What are 5 states generating the maximum and minimum sales revenue?#
## Maximum sales
new_cur=db.cursor()
res=new_cur.execute("select *, round(sum(sales),2) as sales_revenue \
                    from Customer \
                    join Sale on sale.Customer_ID=Customer.Customer_ID join 'Order'\
                    on Customer.Customer_ID='Order'.Customer_ID\
                    join Address\
                    on Address.Order_ID='order'.Order_ID\
                    group by State\
                    order by sales_revenue DESC\
                    limit 5;")
ans=res.fetchall()
df=[]
for i in ans:
    df.append(i)
dff3=pd.DataFrame(df,columns=['Customer_ID','First_Name','Last_Name','Segment','Product_ID','Quantity','Discount','Profit','Sales','Customer_ID','Order_Date','Order_ID','Customer_ID','Country','City','State','Postal_Code','Region','Order_ID','Address_ID','Sales_revenue'])
dff3
```

```python
# 4 What are 5 states generating the maximum and minimum sales revenue?#
## Minimum sales
new_cur=db.cursor()
res=new_cur.execute("select *, round(sum(sales),2) as sales_revenue \
                    from Customer \
                    join Sale on sale.Customer_ID=Customer.Customer_ID join 'Order'\
                    on Customer.Customer_ID='Order'.Customer_ID\
                    join Address\
                    on Address.Order_ID='order'.Order_ID\
                    group by State\
                    order by sales_revenue \
                    limit 5;")
ans=res.fetchall()
df=[]
for i in ans:
    df.append(i)
dff3=pd.DataFrame(df,columns=['Customer_ID','First_Name','Last_Name','Segment','Product_ID','Quantity','Discount','Profit','Sales','Customer_ID','Order_Date','Order_ID','Customer_ID','Country','City','State','Postal_Code','Region','Order_ID','Address_ID','Sales_revenue'])
dff3
```

```python
# 5 What are the 3 products in each product segment with the highest sales?
new_cur=db.cursor()
res=new_cur.execute("select *\
                    from(\
                    select *, \
                    dense_rank() over (PARTITION by segment  order by sales DESC ) as ranking\
                    from Customer\
                    join Sale\
                    on sale.Customer_ID=Customer.Customer_ID\
                    join Product\
                    on sale.Product_ID=Product.Product_ID\
                    group by Product_Name) \
                    where ranking <4;")
ans=res.fetchall()
df=[]
for i in ans:
    df.append(i)

dff4=pd.DataFrame(df,columns=['Customer_ID','First_Name','Last_Name','Segment','Product_ID','Quantity','Discount','Profit','Sales','Customer_ID','Product_ID','Product_Name','Category','Sub_Category','Ranking'])
dff4
```

```python
# 6 Are they the 3 most profitable products as well?
new_cur=db.cursor()
res=new_cur.execute("select *\
                    from(\
                    select *, \
                    dense_rank() over (PARTITION by segment  order by Profit DESC ) as ranking\
                    from Customer\
                    join Sale\
                    on sale.Customer_ID=Customer.Customer_ID\
                    join Product\
                    on sale.Product_ID=Product.Product_ID\
                    group by Product_Name) \
                    where ranking <4;")
ans=res.fetchall()
df=[]
for i in ans:
    df.append(i)

dff5=pd.DataFrame(df,columns=['Customer_ID','First_Name','Last_Name','Segment','Product_ID','Quantity','Discount','Profit','Sales','Customer_ID','Product_ID','Product_Name','Category','Sub_Category','Ranking'])
dff5
```

```python
# 7 What are the 3 best-seller products in each product segment? (Quantity-wise)
# First create a temporary table
new_cur=db.cursor()
res=new_cur.execute("Create TEMPORARY Table if not exists temp1 as \
                    select *, \
                    sum(Quantity) over (PARTITION by Product_Name ) as count_quantity\
                    from Customer\
                    join Sale\
                    on sale.Customer_ID=Customer.Customer_ID \
                    join Product\
                    on sale.Product_ID=Product.Product_ID;")
```

```python
# 7 What are the 3 best-seller products in each product segment? (Quantity-wise)
new_cur=db.cursor()
res=new_cur.execute("select *\
                    from(\
                    select *,\
                    dense_rank() over (PARTITION by segment order by count_quantity DESC ) as ranking\
                    from temp1\
                    group by Product_Name)\
                    where ranking <4\
                    order by segment;")
ans=res.fetchall()
df=[]
for i in ans:
    df.append(i)

dff6=pd.DataFrame(df,columns=['Customer_ID','First_Name','Last_Name','Segment','Product_ID','Quantity','Discount','Profit','Sales','Customer_ID','Product_ID','Product_Name','Category','Sub_Category','Count_Quantity','Ranking'])
dff6
```

```python
# 8 What are the top 3 worst-selling products in every category? (Quantity-wise)
new_cur=db.cursor()
res=new_cur.execute("select *\
                    from(\
                    select *,\
                    dense_rank() over (PARTITION by segment order by count_quantity ) as ranking\
                    from temp1\
                    group by Product_Name)\
                    where ranking <4\
                    order by segment;")
ans=res.fetchall()
df=[]
for i in ans:
    df.append(i)

dff7=pd.DataFrame(df,columns=['Customer_ID','First_Name','Last_Name','Segment','Product_ID','Quantity','Discount','Profit','Sales','Customer_ID','Product_ID','Product_Name','Category','Sub_Category','Count_Quantity','Ranking'])
dff7
```

```python
# 9 How many unique customers per month are there for the year 2016. (There's a catch here: contrary to other 'heavier'
# RDBMS, SQLite does not support the functions YEAR() 
# or MONTH() to extract the year or the month in a date. You will have to create two new columns: year and month.)
new_cur=db.cursor()
res=new_cur.execute("select \
                    count(distinct customer.Customer_ID) as count_cust,\
                    strftime('%m' , order_date) as mnth,\
                    strftime('%Y', order_date) as yr\
                    from 'order'\
                    JOIN Customer\
                    on customer.Customer_ID='order'.customer_ID\
                    WHERE  yr = '2016' \
                    GROUP BY mnth;")
ans=res.fetchall()
df=[]
for i in ans:
    df.append(i)

dff8=pd.DataFrame(df,columns=['Count_Customer','Month','Year'])
dff8
```
