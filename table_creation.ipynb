Connect to the database in sqlite, create a database named Samp.db and read the csv file

```python
import sqlite3
import pandas as pd
db = sqlite3.connect('Samp.db', timeout=10)
dfs = pd.read_csv('Sample-Superstore.csv', sep=',', encoding='ISO-8859-1')
dfs
```

```python
dfs.columns
```

Create ship_ID and Adress_ID

```python
sh_ID=[]
for i in np.arange(0,len(dfs)):
    s = '{}_{}'.format('SH', i)
    sh_ID.append(s)
```

```python
Ad_ID=[]
for i in np.arange(0,len(dfs)):
    s = '{}_{}'.format('AD', i)
    Ad_ID.append(s)
```

```python
dfs['Shipment_ID']=sh_ID
```

```python
dfs['Address_ID']=Ad_ID
```

Replace spaces with underline in the header of columns

```python
dfs.columns=dfs.columns.str.replace(' ','_')
dfs
```

Convert dates from %m/%d/%Y' format to %m-%d-%Y'

```python
dfs['Order_Date']=pd.to_datetime(dfs['Order_Date'], format='%m/%d/%Y')
dfs['Ship_Date']=pd.to_datetime(dfs['Ship_Date'], format='%m/%d/%Y')
```

Split first and last names 

```python
df_n=dfs['Customer_Name'].str.split(' ', n=1, expand=True)
df_n.columns= ['First_Name','Last_Name']
#df_n.set_index(dfs.index)
dfs_n=dfs.join(df_n)
```

create Order table

```python
df1=dfs_n[['Order_ID', 'Order_Date','Customer_ID']]

df1=df1.drop_duplicates()
df1
```

```python
#te=dfs_n.groupby(['Order_ID'])[['Ship_Date', 'Ship_Mode']].sum()
#te
```

Create Shipment table

```python
df2=dfs_n[['Order_ID','Ship_Date', 'Ship_Mode', 'Shipment_ID']]
df2=df2.drop_duplicates()
df2
```

Create Customer table

```python
df3=dfs_n[['Customer_ID', 'First_Name','Last_Name','Segment']]
df3=df3.drop_duplicates()
df3
```

```python
dfs_n.columns
```

Create Product table

```python
df4=dfs_n[['Product_ID','Product_Name','Category','Sub-Category']]
df4.columns=df4.columns.str.replace('-','_')
df4=df4.drop_duplicates()
df4
```

Create Sales table

```python
df5=dfs_n[['Product_ID','Customer_ID','Sales','Quantity','Profit','Discount']]
df5=df5.drop_duplicates()
df5
```

Create Address table

```python
df6=dfs_n[['Country','City','State','Postal_Code','Region','Order_ID','Address_ID']]
df6=df6.drop_duplicates()
df6
```

Populate date to the tables

```python
df1.to_sql('Order',db, index=False,if_exists='append')
df2.to_sql('Shipment',db,index=False, if_exists='append')
df3.to_sql('Customer',db, index=False, if_exists='append')
df4.to_sql('Product',db, index=False, if_exists='append')
df5.to_sql('Sale',db,index=False, if_exists='append')
df6.to_sql('Address',db,index=False, if_exists='append')
```
